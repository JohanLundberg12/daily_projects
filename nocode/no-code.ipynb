```python
from imblearn.datasets import make_imbalance
```

```python
iris = datasets.load_iris(as_frame=True)
```

```python
sampling_strategy = {0: 10, 1: 20, 2: 47}
X, y = make_imbalance(iris.data, iris.target, sampling_strategy=sampling_strategy)
```

```python
import matplotlib.pyplot as plt

target_unique = iris.target.unique()
colors = ["red", "blue", "orange"]
target_colors = [colors[i] for i in target_unique]
y_unique = y.unique()
y_colors = [colors[i] for i in y_unique]

fig, axs = plt.subplots(ncols=2, figsize=(10, 5))
autopct = "%.1f"
iris.target.value_counts().plot.pie(autopct=autopct, ax=axs[0])
axs[0].set_title("Original")
y.value_counts().plot.pie(autopct=autopct, ax=axs[1])
axs[1].set_title("Imbalanced")
fig.tight_layout()
```

```python
import lazypredict
```

```python
from lazypredict.Supervised import LazyClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
data = datasets.load_iris()
X = data.data
y= data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.5,random_state =123)
```

```python
clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(X_train, X_test, y_train, y_test)

print(models)
```
